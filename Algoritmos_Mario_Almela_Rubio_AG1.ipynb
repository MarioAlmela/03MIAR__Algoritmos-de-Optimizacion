```python
# Torres de Hanoi
def torres_hanoi(n, origen, destino, pivote):
  if n == 1:
    print(f'Se mueve el bloque superior del poste {origen} al poste {destino}')
    return
  torres_hanoi(n-1, origen, pivote, destino)
  print(f'Se mueve el bloque superior del poste {origen} al poste {destino}')
  torres_hanoi(n-1, pivote, destino, origen)
```

```python
torres_hanoi(8, 1, 3, 2)
```

```python
# Devolver cambio
def cambio_monedas(cantidad, sistema=[25, 10, 5, 1]):
  print(f'Sistema monetario empleado: \n{sistema}')
  solucion = []
  valor_acumulado = 0

  for moneda in sistema:
    num_monedas = int((cantidad - valor_acumulado)/moneda)
    solucion.append(num_monedas)
    valor_acumulado += num_monedas*moneda

    if valor_acumulado == cantidad:
      return solucion
  
  return solucion
```

```python
cambio_monedas(37)
```
