```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
f = lambda x: x[0]**2 + x[1]**2
df = lambda x: [2*x[0], 2*x[1]]
```

```python
# Preparación de datos para dibujar el mapa de niveles
resolucion = 100
rango = 2.5
X = np.linspace(-rango, rango, resolucion)
Y = np.linspace(-rango, rango, resolucion)
Z = np.zeros((resolucion, resolucion))

for ix, x in enumerate(X):
  for iy, y in enumerate(Y):
    Z[ix, iy] = f([x, y])

# Visualizamos la función
contour = plt.contourf(X, Y, Z, resolucion)
cbar = plt.colorbar(contour)

# Generamos un punto de partida aleatorio y calculamos el gradiente
P = [random.uniform(-rango, rango), random.uniform(-rango, rango)]
plt.plot(P[0], P[1], 'o', c='red')

alpha = 0.05

for _ in range(500):
  grad = df(P)
  P[0], P[1] = P[0] - alpha*grad[0], P[1] - alpha*grad[1]
  plt.plot(P[0], P[1], 'o', c='red')

plt.plot(P[0], P[1], 'o', c='green')
plt.show()
print('Solución: ', P, f(P))
```
